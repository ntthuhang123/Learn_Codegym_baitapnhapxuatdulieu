```python
danh_sach_sp = {}

def them_san_pham(danh_sach_sp):
    print('==Chuc nang them san pham')
    ma_sp = input('Nhap ma san pham: ')
    ten_sp = input('Nhap ten san pham: ')
    if ma_sp in danh_sach_sp:
        print(f'San pham voi ma {ma_sp} da ton tai.')
    else:
        danh_sach_sp[ma_sp] = {
            'ten': ten_sp
        }

        print(f'them san pham thanh cong voi{ma_sp}')

def sua_san_pham(danh_sach_sp):
    print('== Chuc nang sua san pham')
    ma_sp = input("Nhap ma san pham can sua: ")
    if ma_sp not in danh_sach_sp:
        print('San pham khong ton tai')
        return
    ten_sp = input('Nhap ten moi cho san pham: ')
    danh_sach_sp[ma_sp] = {
        'ten': ten_sp
    }
    print(f'Cap nhat thanh cong san pham co ma {ma_sp}')



def xoa_san_pham(danh_sach_sp):
    print('==Chuc nang xoa san pham')
    ma_sp = input('Nhap vao ma san pham can xoa: ')
    if ma_sp not in danh_sach_sp:
        print('San pham khong ton tai')
        return
    danh_sach_sp.pop(ma_sp)
    print(f'Xoa thanh cong san pham voi ma {ma_sp}')



def hien_thi_danh_sach_sp(danh_sach_sp):
    print('Chuc nang hien thi danh sach san pham')
    for (ma_sp,thong_tin) in danh_sach_sp.items():#Cac phan tu la tuple
        print(f'Ma: {ma_sp}. Ten{thong_tin["ten"]}')


while True:
    print('=== Chuong trinh quan li san pham ===')
    print('Chon thao tac: ')
    print('1.Them san pham')
    print('2.Sua san pham')
    print('3.Xoa san pham')
    print('4.Hien thi danh sach san pham')
    print('5.Thoat')
    lua_chon = input('Nhap so thao tac (1-5): ')
    if lua_chon == '1':
        them_san_pham(danh_sach_sp)
    elif lua_chon == '2':
        sua_san_pham(danh_sach_sp)
    elif lua_chon == '3':
        xoa_san_pham(danh_sach_sp)
    elif lua_chon == '4':
        hien_thi_danh_sach_sp(danh_sach_sp)
    elif lua_chon == '5':
        break
    else:
        print('Thao tac khong ton tai')
```
