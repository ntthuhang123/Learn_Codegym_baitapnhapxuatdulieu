```python
import turtle
import math

t = turtle.Turtle()


def drawRectangle(width, height, color):
    t.fillcolor(color)
    t.begin_fill()
    for i in range(2):
        t.forward(width)
        t.left(90)
        t.forward(height)
        t.left(90)
    t.end_fill()


def drawTriangle(length, color):
    t.fillcolor(color)
    t.begin_fill()
    t.forward(length)
    t.left(135)
    t.forward(length / math.sqrt(2))
    t.left(90)
    t.forward(length / math.sqrt(2))
    t.left(135)
    t.end_fill()


def draw_house():
    # vẽ thân nhà
    drawRectangle(150, 80, 'lightblue')
    t.forward(50)
    # vẽ cửa nhà
    drawRectangle(50, 45, 'green')
    t.penup()
    t.goto(-100, 80)
    t.pendown()
    # vẽ mái nhà
    drawTriangle(120, 'brown')
    t.forward(120)
    t.left(135)
    t.forward(120 / math.sqrt(2) / 2)
    t.setheading(90)
    # vẽ ống khói
    t.fillcolor("grey")
    t.begin_fill()
    t.forward(30)
    t.left(90)
    t.forward(10)
    t.left(90)
    t.forward(20)
    t.end_fill()
    t.setheading(90)
    # vẽ khói
    t.penup()
    t.forward(40)
    t.pendown()
    t.forward(20)
    t.penup()
    t.forward(3)
    t.right(90)
    t.forward(5)
    t.right(90)
    t.pendown()
    t.forward(35)


def draw_tree(tanLa, thanCay):
    drawRectangle(thanCay, thanCay * 2, "brown")
    t.setheading(90)
    t.penup()
    # t.goto(65, -90)
    t.forward(thanCay * 2)
    t.left(90)
    t.forward((tanLa - thanCay) / 2)
    t.setheading(0)
    t.pendown()
    drawTriangle(tanLa, "lightgreen")
    t.penup()
    t.setheading(90)
    t.forward(tanLa / 2)
    t.right(90)
    t.forward(5)
    t.pendown()
    drawTriangle(tanLa - 10, "lightgreen")
    t.penup()
    t.setheading(90)
    t.forward((tanLa - 10) / 2)
    t.right(90)
    t.forward(5)
    t.pendown()
    drawTriangle(tanLa - 20, "lightgreen")


t.penup()
t.goto(100, -130)
t.pendown()
draw_tree(90, 20)
t.penup()
t.goto(-100, 0)
t.pendown()
t.fillcolor('blue')
draw_house()
t.penup()
t.goto(-175, -60)
t.pendown()
t.setheading(0)
draw_tree(70, 15)

turtle.done()
```
