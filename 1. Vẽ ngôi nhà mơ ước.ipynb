```python
import turtle
t=turtle.Turtle()

t.penup()
t.goto(-300,-200)
t.pendown()

# vẽ mặt trước ngôi nhà
t.fillcolor("blue")
t.begin_fill()
t.forward(100)
t.left(90)
t.forward(120)
t.left(90)
t.forward(100)
t.left(90)
t.forward(120)
t.end_fill()
# vẽ mặt trước nhà tam giác

t.backward(120)
t.fillcolor("magenta")
t.begin_fill()
t.left(135)
t.forward(70.7)
t.right(90)
t.forward(70.7)
t.end_fill()
# vẽ mái trên
t.fillcolor("Orange")
t.begin_fill()
t.left(70)
t.forward(90)
t.left(110)
t.forward(70.7)
t.left(70)
t.forward(90)
t.end_fill()
#vẽ mặt hông
t.penup()
t.goto(-200,-200)
t.fillcolor("yellow")
t.begin_fill()

t.backward(90)
t.right(115)
t.forward(120)

t.end_fill()

#vẽ cửa chỉnh
t.penup()
t.goto(-280,-200)
t.fillcolor("light green")
t.begin_fill()
t.pendown()
t.forward(90)
t.right(90)
t.forward(40)
t.right(90)
t.forward(90)
t.end_fill()
# vẽ cửa sổ
t.penup()
t.goto(-180,-100)
t.pendown()
t.fillcolor("brown")
t.begin_fill()
t.forward(50)
t.left(115)
t.forward(50)
t.left (65)
t.forward(50)
t.left (115)
t.forward(50)
t.end_fill()
t.penup()
t.goto (-250,100)
t.write("House", font=('Arial', 30, 'italic'), align='right')
# Vẽ cây
#gốc cây
t.penup()
t.goto(50,-250)
t.fillcolor("brown")
t.begin_fill()
t.pendown()
t.right(115)
t.forward(50)
t.right(90)
t.forward(30)
t.right(90)
t.forward(50)
t.right(90)
t.forward(30)
t.end_fill()
# vẽ lá cây1
t.goto(50,-200)
t.fillcolor("light green")
t.begin_fill()
t.forward(50)
t.right(135)
t.forward(91.92)
t.right(90)
t.forward(91.92)
t.right(135)
t.forward(50)
t.end_fill()
# vẽ lá cây 2
t.fillcolor("light green")
t.begin_fill()
t.penup()
t.goto(65,-135)
t.pendown()
t.forward(60)
t.right(135)
t.forward(84.85)
t.right(90)
t.forward(84.85)
t.right(135)
t.forward(60)
t.end_fill()
# vẽ lá cây 3
t.fillcolor("light green")
t.begin_fill()
t.penup()
t.goto(65, -75)
t.pendown()
t.forward(50)
t.right(135)
t.forward(70.71)
t.right(90)
t.forward(70.71)
t.right(135)
t.forward(50)
t.end_fill()
t.penup()
t.goto (150,-250)
t.write("Tree", font=('Arial', 30, 'italic'), align='left')
# vẽ mặt trời
# vẽ hình vuông lớn
t.penup()
t.goto(0, 200)
t.right(180)
t.left(45)
t.forward(100)
t.right(90)
t.forward(100)
t.right(90)
t.forward(100)
t.right(90)
t.forward(100)
t.right(135)
# vẽ đường chéo
t.pendown()
t.forward(141.42)
t.left(135)
t.penup()
t.forward(100)
t.left(135)
t.pendown()
t.forward(141.42)
t.penup()
t.right(135)
t.forward(50)
t.pendown()
t.right(90)
t.forward(100)
t.penup()
t.goto(0, 200)
t.forward(50)
t.pendown()
t.right(90)
t.forward(100)
t.backward(50)
t.right(90)
t.forward(30)
t.left(90)
t.fillcolor("yellow")
t.begin_fill()
t.circle(30)
t.end_fill()
# viết chữ
t.penup()
t.goto(200, 250)
t.write('Sun',font=('Arial', 30, 'italic'), align='left')
# đổ màu nền
t=turtle.Screen()
t.bgcolor("light blue")

turtle.done()
```
